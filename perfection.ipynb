```python
def perfect(a, k, spisok):
    #наш делитель, который накручивается каждую итерацию
    k += 1
    #если делитель делит без остатка, то записываем в наш список
    if a % k == 0:
        spisok.append(k)

    #условия выхода
    #если делитель меньше на 1 (чтобы не выводило само число а)
    if k == a-1:
        return spisok
    #продолжение
    else:
        return perfect(a, k, spisok)
    
```
