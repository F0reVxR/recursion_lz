```python
list1 = []
def convert_to_base(Num, base):
    #находим остаток
    leftover = int(Num % base)

    #целая часть деления без остатка
    intpart = (Num - leftover) / base
    
    if leftover == 10:
        leftover = 'A'
    elif leftover == 11:
        leftover = 'B'
    elif leftover == 12:
        leftover = 'c'
    elif leftover == 13:
        leftover = 'D'
    elif leftover == 14:
        leftover = 'E'
    elif leftover == 15:
        leftover = 'F'

    #переводим в строку
    leftover = str(leftover)
    list1.append(leftover)

    string_list1 = ''.join(list1)
    
    if intpart == 0:
        return string_list1
    else:
        return convert_to_base(intpart, base)
convert_to_base(255, 2)
```
